```python
import requests
from bs4 import BeautifulSoup
```

```python
r = requests.get('https://www.health.ny.gov/statistics/cancer/registry/appendix/neighborhoods.htm')
c = r.content
soup = BeautifulSoup(c, 'html.parser')
soup
```

```python
neighborhood_data = soup.findAll('td', {'headers': 'header2'})
```

```python
neighborhoods = []
for neighborhood in neighborhood_data:
    neighborhoods.append(neighborhood.text.replace(" ","", 1))
    
```

```python
neighborhoods
```

```python
neighborhood_zips = soup.findAll('td', {'headers': 'header3'})
```

```python
neighborhood_zips[0].text.replace(" ", "").split(",")
```

```python
zipcodes = []
for zips in neighborhood_zips:
    zipcodes.append(zips.text.replace(" ", "").split(","))
    
```

```python
zipcodes
```

```python
zipcodes_as_integers = []
for zips in zipcodes:
    integers = [int(i) for i in zips]
    zipcodes_as_integers.append(integers)
        
```

```python
zipcodes_as_integers
```

```python
neighborhoods_and_zipdcodes = {}
for i in range(len(neighborhoods)):
    neighborhoods_and_zipdcodes[neighborhoods[i]] = zipcodes_as_integers[i]
```

```python
neighborhoods_and_zipdcodes
```

```python
from models import Base, Concert, Artist, Genre, Venue
import sqlalchemy
engine = sqlalchemy.create_engine('sqlite:///nycconcertdatabase.db', echo=True)

Base.metadata.create_all(engine)

from sqlalchemy.orm import sessionmaker
Session = sessionmaker(bind=engine)
session = Session()
```

```python
venues = session.query(Venue)
```

```python
venues
```

```python
for venue in venues:
    for key in neighborhoods_and_zipdcodes:
        if venue.zip in neighborhoods_and_zipdcodes[key]:
            venue.neighborhood = key
            session.add(venue)
            session.commit()
```
