```python
from models import Base, Concert, Artist, Genre, Venue
from ticketmasterapi import TicketMasterAPI, TicketmasterEventsParser
from ticketwebscraper import TicketWebParser
from eventbriteapi import EventbriteAPI, EventbriteEventsParser
from bs4 import BeautifulSoup
import sqlalchemy
import requests
import json
engine = sqlalchemy.create_engine('sqlite:///nycconcertdatabase.db', echo=True)

Base.metadata.create_all(engine)

from sqlalchemy.orm import sessionmaker
Session = sessionmaker(bind=engine)
session = Session()
```

```python
venues = session.query(Venue).all()
```

```python
def get_zip_codes():
    zip_codes = {}
    for i in range(5):
        r = requests.get(f"https://app.ticketmaster.com/discovery/v2/events.json?latlong=40.754900,-73.984000&radius=8&unit=miles&source=ticketmaster&classificationName=music&startDateTime=2018-11-09T14:00:00Z&page={i}&size=200&sort=relevance,desc&apikey=rah2o9AS2HbqU4x2DwAbsRZA35MutNT4")
        data = r.json()
        events = data['_embedded']['events']
        for event in events:
            zip_codes[event['_embedded']['venues'][0]['name']] = event['_embedded']['venues'][0]['postalCode']
    return zip_codes
zips = get_zip_codes()
```

```python
def add_zips(venues, zips):
    for venue in venues:
        try:
            venue.zip = zips[venue.name]
            session.add(venue)
            session.commit()
        except:
            pass
add_zips(venues, zips)
```
